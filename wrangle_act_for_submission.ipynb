### Issue #3: Too much columns about the prediction result of dog breed

#### Define
Simplify the table `prediction`:
- create a column `breed` for the prediction result of the dog breed;
- keep the dog breed provided by `p1`, `p2` and `p3` with the highest possibility in the column `breed`;
- if all 3 methods thought the image is not a breed of dog, keep the value in `breed` as "Null";
- drop unnecessary columns.

#### Code

```python
# Create a column `breed` to record the prediction result of dog breed
prediction_clean['breed'] = 'Null'

# This fills in only the rows where p1_dog = True
prediction_clean.loc[prediction_clean.p1_dog, 'breed'] = prediction_clean.loc[prediction_clean.p1_dog, 'p1']

# This fills in only the rows where p1_dog = False but p2_dog = True
prediction_clean.loc[prediction_clean.p2_dog & ~prediction_clean.p1_dog, 'breed'] = prediction_clean.loc[prediction_clean.p2_dog & ~prediction_clean.p1_dog, 'p2']

# This fills in only the rows where p1_dog = False, p2_dog = False but p3_dog = True
prediction_clean.loc[prediction_clean.p3_dog & ~prediction_clean.p1_dog & ~prediction_clean.p2_dog, 'breed'] = prediction_clean.loc[prediction_clean.p3_dog & ~prediction_clean.p1_dog & ~prediction_clean.p2_dog, 'p3']
```

```python
# Drop unwanted columns
prediction_clean = prediction_clean.drop(['p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog'], axis=1)
```
